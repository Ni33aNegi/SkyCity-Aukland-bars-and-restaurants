# Primary objectives

### 1. Quantify total order by channels

```python
total_orders = {
    "Instores": df["InStoreOrders"].sum(),
    "UberEats": df["UberEatsOrders"].sum(),
    "DoorDash": df["DoorDashOrders"].sum(),
    "SelfDelivery": df["SelfDeliveryOrders"].sum()
}
total_orders
```

The above is platform level demand

### 2. Measure channel share distribution across retaurants

```python
df["TotalOrders"] = (
    df["InStoreOrders"] +
    df["UberEatsOrders"] +
    df["DoorDashOrders"] +
    df["SelfDeliveryOrders"]
)

df["InStoreShare_calc"] = df["InStoreOrders"] / df["TotalOrders"]
df["UE_share_calc"] = df["UberEatsOrders"] / df["TotalOrders"]
df["DD_share_calc"] = df["DoorDashOrders"] / df["TotalOrders"]
df["SD_share_calc"] = df["SelfDeliveryOrders"] / df["TotalOrders"]
```

```python
share_distribution = {
    "InStore": df["InStoreShare_calc"].mean(),
    "UberEats": df["UE_share_calc"].mean(),
    "DoorDash": df["DD_share_calc"].mean(),
    "SelfDelivery": df["SD_share_calc"].mean()
}

share_distribution
```

### 3. Identify dominant ordering channles by geography 

```python
geo_channel=df.groupby("Subregion")[[
    "InStoreShare_calc",
    "UE_share_calc",
    "DD_share_calc",
    "SD_share_calc"
]].sum()
```

```python
geo_channel["dominant_channel"] = geo_channel.idxmax(axis=1)
geo_channel
```

# Key performance indicators
1. Channel order shares i.e., volume contribution by channel

```python
df["TotalOrders"] = (df["InStoreOrders"] + df["UberEatsOrders"] + df["DoorDashOrders"] + df["SelfDeliveryOrders"])

for col in["InStore","UberEats","DoorDash","SelfDelivery"]:
    df[f"{col}ShareCheck"] = df[f"{col}Orders"] / df["TotalOrders"]

    # use: identifies dominant channel
    #      compare ordering behviour across segments
```

2. Aggregator dependance index i.e., reliance on third party platforms( UberEats+ DoorDash)

```python
df["AggregatorDependence"]=df["UberEatsOrders"]+df["DoorDashOrders"]/df["TotalOrders"]
# use: Detect margin risk due to commissions
    #  Identify over- dependence on aggregators
```

3. In store reliance ratio i.e., strength of physical store demand


```python
df["InStoreReliance"]=df["InStoreOrders"]/df["TotalOrders"]

# insight: High -> strong brand/ location
#          Low-> over-dependence on delivery
```

4. Channel diversification score i.e., how balanced a retaurant is across channels

```python
def diversity(row):
    shares = row[[
        "InStoreShare",
        "UE_share",
        "DD_share",
        "SD_share"
    ]]
    shares = shares.astype(float)
    shares = shares[shares > 0]
    return -np.sum(shares * np.log(shares))

df["DiversificationScore"] = df.apply(diversity, axis=1)
```

5. Subregion Channel Dominance i.e., Which channel leads in each category

```python
geo_channel=df.groupby("Subregion")[[
    "InStoreOrders",
    "UberEatsOrders",
    "DoorDashOrders",
    "SelfDeliveryOrders"
]].sum()

geo_channel["DominantChannel"]=geo_channel.idxmax(axis=1)

#Insight: Revelasgeographic behaviour
#    helps in regional strategy
```
